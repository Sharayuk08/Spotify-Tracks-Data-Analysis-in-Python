# Set 'release_date' as index

```python
df_tracks.set_index("release_date", inplace = True)
df_tracks.index = pd.to_datetime(df_tracks.index)
df_tracks.head()
```

# Convert song duration to seconds

```python
df_tracks["duration"] = df_tracks["duration_ms"].apply(lambda x : round(x/1000))
```

```python
df_tracks.drop("duration_ms", inplace = True, axis=1)
```

```python
df_tracks.duration.head()
```

# Generate Heat Map

```python
#drop unwanted columns
corr_df = df_tracks.drop(["key", "mode", "explicit"], axis=1).corr(method="pearson")

plt.figure(figsize=(14,6))

heatmap = sns.heatmap(corr_df, annot=True, fmt=".1g", vmin=-1, vmax=1, center=0, cmap="viridis", linewidth=1, linecolor="black")
heatmap.set_title("Correlation Heatmap Between Variables")
heatmap.set_xticklabels(heatmap.get_xticklabels(), rotation=90)
```

# Create Regression Models

## Create sample data which would be 0.4% of the total data

```python
sample_df = df_tracks.sample(int(0.004*len(df_tracks)))
```

```python
print(len(sample_df))
```

### Create Regression plot between loudness and energy

```python
plt.figure(figsize=(10,6))
sns.regplot(data = sample_df, y = "loudness", x = "energy", color = "c").set(title = "Loudness vs Energy Correlation")
```

#### We can see here that there is a high positive correlation between the loudness and energy. 

### Create Regression plot between popularity and acousticness

```python
plt.figure(figsize=(10,6))
sns.regplot(data = sample_df, y = "popularity", x = "acousticness", color = "b").set(title = "Popularity vs Acousticness Correlation")
```

#### From the downward trend of the regression line we can say that there is negative correlation between the popularity and acousticness of the songs. If the acousticness of the song increases, the popularity decreases

## Create new column called year using release_date column

```python
df_tracks['dates']= df_tracks.index.get_level_values('release_date')
df_tracks.dates = pd.to_datetime(df_tracks.dates)
years = df_tracks.dates.dt.year
```

```python
df_tracks.head()
```

```python
sns.displot(years,discrete=True, aspect=2, height=5, kind="hist").set(title= "Number of songs per year")
```

## Duration of songs over the years

```python
total_dr =  df_tracks.duration
fig_dims = (18,7)
fig, ax = plt.subplots(figsize = fig_dims)
fig = sns.barplot(x= years, y = total_dr, ax = ax, errwidth = False).set(title = "Year vs Duration")
plt.xticks(rotation=90)
```

#### Earlier in 1920s the songs had shorter duration. Later, in 1930s the durations were increased. After 2010 we can see that the duration of the songs started decreasing
